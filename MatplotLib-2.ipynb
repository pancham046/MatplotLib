#### Bar Chart

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)
plt.yticks([0,1,2,3,4])
bars[0].set_hatch('/')
plt.figure(figsize = (6,4), dpi = 300)
plt.show()
```

```python
labels = ['A','B','C']
values = [1,4,2]

bars = plt.bar(labels, values)
plt.yticks([0,1,2,3,4])
#bars[0].set_hatch('/')
patterns = ['/','O','*']

for bar in bars:
    bar.set_hatch(patterns.pop(0))

plt.figure(figsize = (6,4), dpi = 300)
plt.show()
```

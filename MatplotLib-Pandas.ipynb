#### Histogram

```python
labels = ['Right', 'Left']

plt.pie(fifa['Preferred Foot'].value_counts(), labels = labels, autopct='%.2f %%')

plt.title('Foot Preference of FIFA Players')

# plt.legend(fifa['Preferred Foot'].unique())

plt.show()
```

```python
fifa['Weight'].head()  # The weights contains the string lbs and hence we need to get rid of them
```

```python
fifa['Weight'] = [x.strip('lbs') if type(x)==str else x for x in fifa['Weight']]
```

```python
fifa['Weight'].head()[0]
```

```python
fifa['Weight'].astype('float32')
```

```python
# # light = fifa[fifa['Weight'] < '125']
# # light

# for Weight in fifa['Weight']:
#     if fifa['Weight'] < 125:
#         fifa['Weight Category'] = 'Light'
#     elif fifa['Weight'] >= 125 & fifa['Weight'] <=175:
#         fifa['Weight Category'] = 'Medium'
#     elif fifa['Weight'] > 175:
#         fifa['Weight Category'] = 'Heavy'


weight_conditions = [(fifa['Weight'].astype('float') < 125),
                    (fifa['Weight'].astype('float') >= 125) & (fifa['Weight'].astype('float') <= 175),
                    (fifa['Weight'].astype('float') > 175)]

weight_category = ['Light','Medium','Heavy']

fifa['Weight Category'] = np.select(weight_conditions, weight_category)
```

```python
fifa['Weight Category'].head()
```

```python
fifa['Weight Category'].value_counts()
```

```python
labels = ['Medium','Heavy','NaN','Light']

explode = [0,0,.2,.2]

plt.pie(fifa['Weight Category'].value_counts(), labels = labels, autopct='%.2f', explode=explode)

plt.show()
```
